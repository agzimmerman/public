```python
from math import pi
from numpy import linspace, sin, sqrt, meshgrid, multiply
from matplotlib.pyplot import plot, figure, cm, show
%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D
```

## Plot the lowest two energy states for the 1D particle in a box problem.

```python
length = 1e-9 # m
x = linspace(0, length, 1e2)
n = 1
Psi = sqrt(2/length)*sin(n*pi*x/length)
plot(x, Psi)
n = 2
Psi = sqrt(2/length)*sin(n*pi*x/length)
plot(x, Psi)
```

## Plot the lowest three energy states for the 2D particle in a box problem.

```python
lengthx = length
lengthy = 2*length
y = linspace(0, lengthy, 1e2)
x, y = meshgrid(x, y)
nx = 1
ny = 1
Psi = 2/sqrt(lengthx**2 + lengthy**2)*multiply(sin(pi*x*nx/lengthx), sin(pi*y*ny/lengthy))
```

```python
fig = figure(figsize=(8,12))
ax1 = fig.add_subplot(3,1,1,projection='3d')
surf = ax1.plot_surface(x, y, Psi, label = 'nx = 1, ny = 1', rstride = 1, cstride = 1, cmap=cm.hot, linewidth = 0)
```
